### A.Web Scraping

#### 1.Pengambilan Data

```python

# URL yang akan diakses
url = 'https://www.tokopedia.com/search?navsource=&page=1&q=seblak&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&st='
driver = webdriver.Chrome()
driver.get(url)
time.sleep(0.5)

# Membuat objek BeautifulSoup untuk parsing HTML halaman web
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
for div_tag in soup.find_all('div',{"class":"prd_link-product-name css-3um8ox"}):
    print(div_tag.get_text())
```

#### 2.Load data `Nama Produk`, `Harga Produk`, `Penjual`, `Kota Toko`, `Banyaknya Terjual`, dan `Rating Produk`.

```python
# Inisialisasi WebDriver (pastikan driver sesuai dengan versi browser yang Anda miliki)
driver = webdriver.Chrome()

# Inisialisasi list untuk menyimpan data
data = []

# Loop melalui halaman pencarian Tokopedia
for i in range(1, 11):
    url = "https://www.tokopedia.com/search?navsource=&page={}&q=seblak&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&st=".format(i)
    driver.get(url)
    time.sleep(1)
    soup = BeautifulSoup(driver.page_source, "html.parser")

    # Loop melalui setiap barang pada halaman
    for barang in soup.find_all('div', {"class": "css-1asz3by"}):
        try:
            nama = barang.find('div', {"class": "prd_link-product-name css-3um8ox"}).text
        except AttributeError:
            nama = None

        try:
            harga = barang.find('div', {"class": "prd_link-product-price css-h66vau"}).text
        except AttributeError:
            harga = None

        try:
            penjual = barang.find('span', {"class": "prd_link-shop-name css-1kdc32b flip"}).text
        except AttributeError:
            penjual = None

        try:
            lokasi = barang.find('span', {"class": "prd_link-shop-loc css-1kdc32b flip"}).text
        except AttributeError:
            lokasi = None

        try:
            terjual = barang.find('span', {"class": "prd_label-integrity css-1sgek4h"}).text
        except AttributeError:
            terjual = None

        try:
            rating = barang.find('span', {"class": "prd_rating-average-text css-t70v7i"}).text
        except AttributeError:
            rating = None

        # Menambahkan data ke dalam list
        data.append((nama, harga, penjual, lokasi, terjual, rating))

# Menutup WebDriver setelah selesai
driver.quit()

# Membuat DataFrame dari list data
df = pd.DataFrame(data, columns=['Nama_Produk', 'Harga_Produk', 'Penjual', 'Kota_Toko', 'Banyaknya_Terjual', 'Rating_Produk'])
df
```

#### 2.Data Cleaning

sebelum cleaning data, pertama saya cek dulu didalam datanya apa saja hal yang harus dihapus atau tidak dipakai..

```python
df.Banyaknya_Terjual.unique()
```

```python
# Menghapus string 'terjual ' dan mengganti string 'rb' dengan *1000 di kolom 'Banyaknya_Terjual'
df['Harga_Produk'] = df['Harga_Produk'].replace({'Rp': ''}, regex=True)

df['Banyaknya_Terjual'] = df['Banyaknya_Terjual'].replace({'terjual': '',' ': '', '\+': '', 'rb': '000'}, regex=True)

df['Banyaknya_Terjual'].fillna(0, inplace=True)

# Menampilkan DataFrame setelah perubahan
df
```

```python
# Menyesuaikan tipe data kolom 'Harga_Produk' menjadi float (jika belum)
df['Harga_Produk'] = df['Harga_Produk'].astype(float)

# Menyesuaikan tipe data kolom 'Banyaknya_Terjual' menjadi int (jika belum)
df['Banyaknya_Terjual'] = df['Banyaknya_Terjual'].astype(int)

# Menyesuaikan tipe data kolom 'Rating_Produk' menjadi float (jika belum)
df['Rating_Produk'] = df['Rating_Produk'].astype(float)
```

#### 3.Hypothesis Test

di hipotesis saya akan mengklasisfikasi berdasarkan jabodetabek dan non jabodetabek

```python
jabodetabek = data[(data["Kota_Toko"]=="Jakarta Pusat") | (data["Kota_Toko"]=="Jakarta Selatan") | (data["Kota_Toko"]=="Depok") | (data["Kota_Toko"]=="Tangerang Selatan") | (data["Kota_Toko"]=="Jakarta Barat") | (data["Kota_Toko"]=="Jakarta Timur") | (data["Kota_Toko"]=="Tangerang")]

jabodetabek
```

```python
non_jabodetabek = data[~((data["Kota_Toko"]=="Jakarta Pusat") | (data["Kota_Toko"]=="Jakarta Selatan") | (data["Kota_Toko"]=="Depok") | (data["Kota_Toko"]=="Tangerang Selatan") | (data["Kota_Toko"]=="Jakarta Barat") | (data["Kota_Toko"]=="Jakarta Timur") | (data["Kota_Toko"]=="Tangerang"))]

non_jabodetabek
```

sekarang saya akan menganalisis harga di jabodetabek dan diluar jabodetabek.

Di Hypothesis Test saya akan menggunakan `Two-Sample Independent Test`. ddimana akan mengetes dua kondisi. 

```python
jabodetabek_areas = ["Jakarta Pusat", "Jakarta Selatan", "Depok", "Tangerang Selatan", "Jakarta Barat", "Jakarta Timur", "Tangerang"]
jabodetabek = data[data["Kota_Toko"].isin(jabodetabek_areas)]
jabodetabek_total = jabodetabek['Harga_Produk'].mean()
jabodetabek_total
```

```python
jabodetabek_areas = ["Jakarta Pusat", "Jakarta Selatan", "Depok", "Tangerang Selatan", "Jakarta Barat", "Jakarta Timur", "Tangerang"]
non_jabodetabek = data[~data["Kota_Toko"].isin(jabodetabek_areas)]
non_jabodetabek_total = non_jabodetabek['Harga_Produk'].mean()
non_jabodetabek_total
```

```python
t_stat, p_val = stats.ttest_ind(jabodetabek['Harga_Produk'],non_jabodetabek['Harga_Produk'])
print('T-Statistic:',t_stat)
print('P-value:',p_val)
```

berdasarkan data yang di dapat disimpulkan bahwa :

H0: Harga barang di jabodetabek == harga barang di luar jabodetabek

H1: Harga barang di jabodetabek != harga barang di luar jabodetabek

#### 4.Satisfaction

disini akan menghitung satisfaction dan korelasi 

```python
corr_r, pval_p = stats.pearsonr(df['Harga_Produk'], df['Banyaknya_Terjual'])
corr_rho, pval_s = stats.spearmanr(df['Harga_Produk'], df['Banyaknya_Terjual'])
corr_tau, pval_k = stats.kendalltau(df['Harga_Produk'], df['Banyaknya_Terjual'])

print(f"r-correlation: {corr_r:.2f}, p-value: {pval_p}")
print(f"rho-correlation: {corr_rho:.2f}, p-value: {pval_s}")
print(f"tau-correlation: {corr_tau:.2f}, p-value: {pval_k}")
```

dari hasil perhitungan hasil diatas korelasi dari harga produk dan banyaknya terjual bersifal by chance karena nilai p-value kurang dari 0.05
